### Question 1: OLAP Cube: Write a query to create the following OLAP cube

### Revenue of each movie by Title, Month, Store City and Customer City [5 Marks]


```python
%%time
%%sql

SELECT 
    m.title,
    d.month, 
    s.city as store_city,
    c.city as customer_city,
    sum(f.sales_amount) AS revenue
FROM factSales f
JOIN dimDate d  ON f.date_key  = d.date_key
JOIN dimMovie m ON m.movie_key = f.movie_key
JOIN dimStore s ON s.store_key = f.store_key
JOIN dimCustomer c ON c.customer_key = f.customer_key
GROUP BY (m.title, d.month, s.city, c.city)
ORDER BY revenue DESC
LIMIT 20;
```

### Question 2: Roll Up: Write a query to Roll Up the above created cube by Customer Country [5 Marks]

```python
%%time
%%sql

SELECT 
    m.title,
    d.month, 
    s.city as store_city,
    c.country as customer_country,
    sum(f.sales_amount) AS revenue
FROM factSales f
JOIN dimDate d  ON f.date_key  = d.date_key
JOIN dimMovie m ON m.movie_key = f.movie_key
JOIN dimStore s ON s.store_key = f.store_key
JOIN dimCustomer c ON c.customer_key = f.customer_key
GROUP BY (m.title, d.month, s.city, c.country)
ORDER BY revenue DESC
LIMIT 20;
```

### Question 3: Drill Down: Write a query to Drill Down the above created cube by Store Districts [5 Marks]

```python
%%time
%%sql

SELECT 
    m.title,
    d.month, 
    s.district as store_district,
    c.city as customer_city,
    sum(f.sales_amount) AS revenue
FROM factSales f
JOIN dimDate d  ON f.date_key  = d.date_key
JOIN dimMovie m ON m.movie_key = f.movie_key
JOIN dimStore s ON s.store_key = f.store_key
JOIN dimCustomer c ON c.customer_key = f.customer_key
GROUP BY (m.title, d.month, s.district, c.city)
ORDER BY revenue DESC
LIMIT 20;
```

### Question 4: Slicing: Write a query to extract a slice of cube (created in question 1) for the 3rd month [5 Marks]

```python
%%time
%%sql

SELECT 
    m.title,
    d.month, 
    s.district as store_district,
    c.city as customer_city,
    sum(f.sales_amount) AS revenue
FROM factSales f
JOIN dimDate d  ON f.date_key  = d.date_key
JOIN dimMovie m ON m.movie_key = f.movie_key
JOIN dimStore s ON s.store_key = f.store_key
JOIN dimCustomer c ON c.customer_key = f.customer_key
WHERE d.month = 3
GROUP BY (m.title, d.month, s.district, c.city)
ORDER BY revenue DESC
LIMIT 20;
```

### Question 5: Dicing: Write a query to perform dicing on cube (created in question 1) on the following conditions [5 Marks]

* months can be 2,3,4
* ratings can be PG-13, PG, G
* store city in Lethbridge or Woodridge
* customer city in Mukateve, Yingkou, San Bernardino, Syrakusa

```python
%%time
%%sql

SELECT
    m.rating,
    m.title,
    d.month, 
    s.city as store_city,
    c.city as customer_city,
    sum(f.sales_amount) AS revenue
FROM factSales f
JOIN dimDate d  ON f.date_key  = d.date_key
JOIN dimMovie m ON m.movie_key = f.movie_key
JOIN dimStore s ON s.store_key = f.store_key
JOIN dimCustomer c ON c.customer_key = f.customer_key
WHERE m.rating IN ('PG-13', 'PG','G')
AND s.city IN ('Woodridge','Lethbridge')
AND c.city IN ('Mukateve', 'Yingkou','San Bernardino', 'Syrakusa')
AND d.month IN (2, 3, 4)
GROUP BY (m.title,m.rating, d.month, s.city, c.city)
ORDER BY revenue DESC
LIMIT 20;
```

### Question 6: Write a query to find out the total revenue, revenue by month, customer city, revenue by month and customer city [10 Marks]

```python
%%sql
SELECT d.month, c.city, sum(sales_amount) as revenue
FROM factSales f
JOIN dimDate d ON f.date_key = d.date_key
JOIN dimStore s ON f.store_key = s.store_key
JOIN dimCustomer c ON c.customer_key = f.customer_key
GROUP BY grouping sets ((), d.month, c.city, (d.month, c.city))
```

### Question 7: Find out the revenue generated by all of the movies that were sold in Woodridge (Store City) that were released in the year 1997 [5 Marks]

```python
%%time
%%sql

SELECT
    m.title,
    sum(f.sales_amount) AS revenue
FROM factSales f
JOIN dimDate d  ON f.date_key  = d.date_key
JOIN dimMovie m ON m.movie_key = f.movie_key
JOIN dimStore s ON s.store_key = f.store_key
JOIN dimCustomer c ON c.customer_key = f.customer_key
WHERE
s.city  = 'Woodridge'
AND d.year = 1997
GROUP BY (m.title)
ORDER BY revenue DESC
LIMIT 20;
```

### Question 8: Find out the top 3 movies sold/rented from the store in Lethbridge (Store City) w.r.t revenue given the movie rating was PG-13 [5 Marks]

```python
%%time
%%sql

SELECT
    m.title,
    sum(f.sales_amount) AS revenue
FROM factSales f
JOIN dimDate d  ON f.date_key  = d.date_key
JOIN dimMovie m ON m.movie_key = f.movie_key
JOIN dimStore s ON s.store_key = f.store_key
JOIN dimCustomer c ON c.customer_key = f.customer_key
WHERE m.rating ='PG-13'
AND s.city ='Lethbridge'
GROUP BY (m.title)
ORDER BY revenue DESC
LIMIT 3;
```

### Question 9: Find out the top 10 Customers who spent the most money on 'PG' rated movies [5 Marks]

```python
%%time
%%sql

SELECT
    c.customer_key,
    sum(f.sales_amount) AS money_spent
FROM factSales f
JOIN dimDate d  ON f.date_key  = d.date_key
JOIN dimMovie m ON m.movie_key = f.movie_key
JOIN dimStore s ON s.store_key = f.store_key
JOIN dimCustomer c ON c.customer_key = f.customer_key
WHERE m.rating = 'PG'
GROUP BY (c.customer_key)
ORDER BY money_spent DESC
LIMIT 10;
```
